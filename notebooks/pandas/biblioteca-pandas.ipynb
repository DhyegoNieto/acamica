### concat y append 

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
   

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])


df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
frames = [df1, df2, df3]
```

```python
df1
```

```python
pd.concat(frames)
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                     index=[2, 3, 6, 7])
df4
```

```python
pd.concat([df1, df4], axis=0) # Concatenación horizontal
```

```python
pd.concat([df1, df4], axis=1, join='inner') # Cambiando la forma de concatenar con join
```

# Concat

### Note how the autogenerated DataFrame indexes doesn't have an order

```python
df_concat = pd.concat([df1, df2])
df_concat
```

### Adding the parameter **ignore_index=True** will set an order for the indexes

```python
df_concat = pd.concat([df1, df2], ignore_index=True)
df_concat
```

### Also you can set a label to teh DataFrame to indentify which data came from which DataFrame
### this is done using the additional parameter **keys**

```python
df_keys = pd.concat([df1, df2], keys=['df1', 'df2'])
df_keys
```

### Using this labels will make easier to locate the data coming from a specific DataFrame

```python
df_keys.loc['df1']
```

### Concatenate DataFrames along column, you can specify the parameter **axis = 1**

```python
df_col = pd.concat([df1, df2], axis=1)
df_col
```

### bins

```python
# Creates a DataFrame with random numbers
df = pd.DataFrame(np.random.randint(0, 100,size=(100, 4)), columns=list('ABCD'))
df.head()
```

```python
pd.qcut(range(5), 4).value_counts()
```

```python
pd.qcut(range(5), 3, labels=['good', 'medium', 'bad']).value_counts()
```

- **qcut will creates the bins with the most equal distribution, i.e. if there are 30 values for 3 bins will assign 10 values for each bin**

- **In the other hand cut will distribute the values according the value of the frequency, 
higher frequencies in the inner bins and fewer in the outer, essentially a tabular form of a histogram (bell shaped)**

```python
pd.qcut(np.random.randn(30), 3).value_counts()
```

```python
pd.cut(np.random.randn(30), 3).value_counts()
```
