# Actividad Dataset Iris

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv('./../../dataset/iris.csv')
```

```python
sns.set(style="ticks", color_codes=True)
```

```python
sns.distplot(iris['SepalLengthCm'], bins=None, hist=True, kde=True, rug=False)
```

```python
sns.pairplot(iris, hue='Species', size=3.0)
```

# Correlación SepalLengthCm & PetalLengthCm

```python
iris['SepalLengthCm'].corr(iris['PetalLengthCm'])
```

```python
sns.pairplot(iris, x_vars=['SepalLengthCm'], y_vars=['PetalLengthCm'], height=6.0)
```

```python
sns.lineplot(x=np.linspace(0,10,150), y=iris['SepalWidthCm'], data=iris, label='Sepal Width')
sns.lineplot(x=np.linspace(0,10,150), y=iris['PetalLengthCm'], data=iris, color='r', label='Petal Length')
```
