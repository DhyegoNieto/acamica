# Demo Subplots

** matplotlib.pyplot.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, fig_kw) **

![title](./../../images/matplotlib/subplots.png)

```python
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
plt.close('all')
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple Plot')
```

```python
fig, axarr = plt.subplots(2, sharex=True)
fig.suptitle('Sharing X Axis')
axarr[0].plot(x,y)
axarr[1].scatter(x, y)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
fig.suptitle('Sharing Y Axis')
ax1.plot(x, y)
ax2.scatter(x, y)
```

## Three subplots sharing both x/y axes

```python
fig, axarr = plt.subplots(3, 1, sharex=True, sharey=True)
fig.suptitle('Sharing Both Axes')
axarr[0].plot(x, y, marker='+')
axarr[1].scatter(x, y)
axarr[2].scatter(x, 2 * y ** 2 - 1, color='r', marker='^')
#Bring subplots close to each other
#fig.subplots_adjust(hspace=0)
#hide x labels and tick label for all but bottom plot.

#for ax in axarr:
    #ax.label_outer()
```

## Row and Column Sharing

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='y', marker='.')
ax4.plot(x, 2 * y ** 2 -1, color='r')
```

## Four axes

```python
fig, axarr = plt.subplots(2, 2)
axarr[0, 0].plot(x, y)
axarr[0, 0].set_title('Axis [0, 0]')
axarr[0, 1].scatter(x, y)
axarr[0, 1].set_title('Axis [0, 1]')
axarr[1, 0].plot(x, y ** 2)
axarr[1, 0].set_title('Axis[1, 0]')
axarr[1, 1].scatter(x , y ** 2)
axarr[1, 1].set_title('Axis [1, 1]')
for ax in axarr.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
# Hide x labels and tick labels for top plots and y ticks for right plots
for ax in axarr.flat:
    ax.label_outer()
```

```python
fig, axarr = plt.subplots(2, 2, subplot_kw=dict(projection='polar'))
axarr[0, 0].plot(x,y)
axarr[0, 0].set_title('Axis [0, 0]')
axarr[0, 1].scatter(x,y)
axarr[0, 1].set_title('Axis [0, 1]')
axarr[1, 0].plot(x, y ** 2)
axarr[1, 0].set_title('Axis [1, 0]')
axarr[1, 1].scatter(x, y ** 2)
axarr[1, 1].set_title('Axis [1, 1]')
```

# Histogram

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# Create some data to use for the plot
dt = 0.001
# Start, Stop, Step
t = np.arange(0., 10.0, dt)
# calculate the exponential of all elements in the input array
r = np.exp(-t[:1000] / 0.05) # impulse response
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)] * dt # Colored Noise
```

### Main Subplot

```python
plt.plot(t, s)
plt.axis([0, 1, 1.1 * np.min(s), 2 * np.max(s)])
plt.xlabel('Time (s)')
plt.ylabel('Current (nA)')
plt.title('Gaussian Colored Noise')

# this is an inset axes over the main axes
a = plt.axes([.65, .6, .2, .2], facecolor='k')
n, bins, patches = plt.hist(s, 400, normed=True)
plt.title('Probability')
plt.xticks([])
plt.yticks([])

# this is another inset axes over the main axes
a = plt.axes([0.2, 0.6, .2, .2], facecolor='k')
plt.plot(t[:len(r)], r)
plt.title('Impulse response')
plt.xlim(0, 0.2)
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# The histogram of the data
n, bis, patches = plt.hist(x, 50, normed=True, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
