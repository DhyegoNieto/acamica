```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_original = pd.read_csv("winemag-data_first150k.csv", sep = ";")
```

```python
df_original.shape
```

```python
df_original.head()
```

```python
# Bar plot
df_original['country'].value_counts().head(10).plot.bar()
```

```python
df_original['points'].value_counts().sort_index().plot.bar()
```

```python
# Line Plot
df_original['price'].value_counts().sort_index().plot.line()
```

```python
df_original.query('price < 100').price.value_counts().sort_index().plot.line()
```

```python
# Area Plot
df_original['points'].value_counts().sort_index().plot.area()
```

```python
# Scatter Plot
df_original.plot.scatter(x='points', y='price')
```

```python
# Parámetros estéticos
df_original['points'].value_counts().sort_index().plot.bar(
    figsize = (10,5), # Permite cambiar el tamaño del gráfico
    fontsize = 10, # Permite cambiar el tamaño de la fuente
    title = "Count of Wine-Points" # Permite colocar un título al gráfico
)
```

# Usando la librería *Seaborn*

```python
sns.countplot(df_original['country'])
```

```python
sns.countplot(df_original['country'], order = df_original.country.value_counts().iloc[:5].index)
```

```python
# Diagrama de densidad
sns.kdeplot(df_original.query('price < 100').price)
```

```python
# Histograma
ax = sns.distplot(df_original['points'], bins = 30, kde = False)
```

```python
# Histograma
ax = sns.distplot(df_original['points'], bins = 8, kde = True)
## A bins se le puede pasar un range bins=range(30,100,5)
```

```python
# Gráfico Box Plot
df1 = df_original[df_original.variety.isin(df_original.variety.value_counts().head(5).index)] # Traigo solo las 5 mayores variedades

plt.figure(figsize=(12, 8)) # Defino el tamaño del gráfico

sns.boxplot(
    x = 'variety',
    y = 'points',
    data = df1
)
```

```python
sns.pairplot(df_original)
```

```python
# Multivariant 
sns.lmplot( x = 'price',
            y = 'points',
            data = df_original)
```
