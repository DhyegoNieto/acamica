## Import dataset

```python
data = pd.read_csv('./../../dataset/telcoChurn.csv')
data.head(3)
```

## Imputing null values

**Replacing null values for -1 for this case**

```python
data['TotalCharges'] = data['TotalCharges'].replace(' ', -1).astype(float)
```

## Turning categorical variables into Dummies

```python
cat_var = ['gender', 'Partner', 'Dependents', 'PhoneService','MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
           'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
           'PaymentMethod']
```

```python
categorical_df = data[cat_var]
categorical_df.head(3)
```

```python
data = data.drop(cat_var, axis=1)
```

```python
for var in cat_var:
    dummy = pd.get_dummies(categorical_df[var], prefix=var)
    data1 = data.join(dummy, lsuffix='left')
data = data1
data.head(3)
```
